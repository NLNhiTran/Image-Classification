```python
import numpy as np

# load train and test data

with np.load('train_data_label.npz') as data: 
    train_data = data['train_data']
    train_label = data['train_label']
    
with np.load('test_data_label.npz') as data: 
    x_test = data['test_data']
    y_test = data['test_label']
    
```

```python
def show_image_by_range(img, labels, start=0, end=1, img_size=28):
    """
        Show images in range from start to end.
    """
    import matplotlib.pyplot as plt

    img_reshaped = img.reshape(-1, img_size, img_size, 1)
    plt.figure(figsize=(15, 15))

    for i in range(start, end):
        plt.subplot((end - start) // 5 + 1, 5, i + 1)
        plt.axis('off')
        plt.title(labels[i])
        plt.imshow(img_reshaped[i, :, :, 0], cmap = 'gray')

    plt.show()
```

```python
## Split the data
X_train, X_val, y_train, y_val = train_test_split(train_data, train_label, 
                                                  test_size=1/3, random_state=123)

label_enc = LabelEncoder()
y_train = label_enc.fit_transform(y_train)
y_val = label_enc.fit_transform(y_val)
```
